```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def check(a):
    f1=False
    f2=False
    for i in a:
        if i.isdigit():
            f1=True
        if i.isalpha():
            f2=True
    return f1 and f2
```

```python
l=[]
for i in df['cost_2']:
    if type(i)==float:
        i=i/2
        l.append(i)
    else:
        x=check(i)
        if x==False:
            if i.isdigit():
                i=float(i)
                i=i/2
                l.append(i)
            else:
                i=np.nan
                l.append(i)
        else:
            i=np.nan
            l.append(i)
```

```python
df['cost']=l
df.drop('cost_2',axis=1,inplace=True)
```

```python
def avg(a):
    x=0
    a=list(a)
    for i in a:
        if i==np.nan:
            pass
        else:
            x+=i
    y=x/len(a)
    return y
```

```python
a=df.groupby('name_new')['rate_5'].unique()
a=pd.DataFrame(a)
l=[]
for i in a['rate_5']:
    i=list(i)
    x=0
    for j in i:
        x+=j
    b=x/len(i)
    l.append(b)
    
        
```

```python
a['count']=a['rate_5'].apply(lambda x:len(x))
a['avg_rate']=l
c=a.sort_values(by='count',ascending=False).head(20)
c
```
